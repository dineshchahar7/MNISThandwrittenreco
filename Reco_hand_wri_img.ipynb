```python
data = df.values
```

```python
# Separating features,label in the dataset (i.e, X and Y)
```

```python
X = data[:,1:]
Y = data[:,:1]
```

```python
# Spliting the dataset into Train and Test datasets.
XT,Xt,YT,Yt = train_test_split(X,Y,test_size= 0.2)
# (XT,YT) --- Test dataset , (Xt,Yt) --- Train dataset
```

```python
# visualising the some data points.

def Imgsho(sample):
    img = sample.reshape(28,28)
    plt.imshow(img,cmap='gray')
    plt.show()
```

```python
# calculating Euclidean distance

def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

# definig KNN

def KNN(X,Y,querypoint,k=11):
    vals =[]
    m = X.shape[0]
    for i in range(m):
        d = dist(querypoint,X[i])
        vals.append((d,Y[i]))
    vals = sorted(vals)
    # Nearest first K points.
    vals = vals[:k]
    vals = np.array(vals,dtype=tuple)
    new_vals = np.unique(vals[:1],return_counts= True)
    
    # print new_values

    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    return pred
```

```python
# Make prediction

prediction = KNN(XT,YT,Xt[4])
print(prediction)
Imgsho(Xt[4])
print(Yt[4])
```
